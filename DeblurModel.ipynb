```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from skimage.data import shepp_logan_phantom

def create_blurred_phantom(size, sigma=2):
    """
    Generates a Shepp-Logan phantom and applies Gaussian blur.
    
    Parameters:
    - size: Tuple of ints, the size of the phantom image.
    - sigma: Float, the standard deviation of the Gaussian kernel.
    
    Returns:
    - phantom: The original Shepp-Logan phantom image.
    - blurred: The Gaussian blurred phantom image.
    """
    # Generate the Shepp-Logan phantom
    phantom = shepp_logan_phantom()
    #phantom = np.pad(phantom, ((0, size[0] - phantom.shape[0]), (0, size[1] - phantom.shape[1])), 'constant', constant_values=0)
    #phantom = phantom[:size[0], :size[1]]  # Ensure the image is exactly the desired size
    
    # Apply Gaussian blur
    blurred = gaussian_filter(phantom, sigma=sigma)
    
    return phantom, blurred

# Parameters
size = (65, 65)  # Image size
sigma = 2  # Standard deviation for Gaussian blur

# Create the phantom and blurred images
phantom, blurred = create_blurred_phantom(size, sigma)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(phantom, cmap='gray')
axes[0].set_title('Original Shepp-Logan Phantom')
axes[0].axis('off')

axes[1].imshow(blurred, cmap='gray')
axes[1].set_title('Gaussian Blurred Phantom')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
def gaussian_set(phantom,n,sigma):
    gaussian_slice = np.zeros((phantom.shape[0], phantom.shape[1]))
    gaussian_stack = np.zeros((n, gaussian_slice.shape[0], gaussian_slice.shape[1]))
    for i in range(n):   
        gaussian_stack[i] = gaussian_filter(phantom, sigma = sigma)
        phantom = gaussian_stack[i]

    return gaussian_stack

test_phantom = gaussian_set(shepp_logan_phantom(), 10, 1)

print(test_phantom)
plt.figure(figsize=(6, 6))
plt.imshow(test_phantom[8], cmap='gray')
plt.title('Original Shepp-Logan Phantom')
plt.axis('off')

print(test_phantom)
plt.figure(figsize=(6, 6))
plt.imshow(test_phantom[1], cmap='gray')
plt.title('Original Shepp-Logan Phantom')
plt.axis('off')
```
